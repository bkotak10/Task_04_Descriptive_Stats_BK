```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
```

### Loading the Facebook ads dataset for analysis
In this section, we use the Pandas library to perform descriptive statistical analysis on datasets related to the 2024 U.S. Presidential Election. Each dataset contains social media activity (ads, posts, tweets) and includes both numeric and categorical data.

We perform the following steps:
	•	Load and preview the dataset
	•	Generate basic descriptive statistics (count, mean, std, min, max, etc.)
	•	Explore column-wise uniqueness and most frequent values
	•	Group the data by page_id and ad_id to understand behavioral patterns at different granularities
	•	Use .value_counts() to investigate the distribution of categorical features such as page_id, gender, or platform

The goal is to uncover structural insights and identify high-level trends that can later inform further exploratory analysis or visualization.

```python
df_ads = pd.read_csv("2024_fb_ads_president_scored_anon.csv", low_memory=False)
df_ads.shape, df_ads.head(3)
```

## Full descriptive stats for facebook ads

```python
df_ads.describe(include='all')
```

## Numeric-only stats

```python
df_ads.select_dtypes(include="number").describe().T
```

## Unique counts in non-numeric column

```python
df_ads.select_dtypes(exclude="number").nunique().sort_values(ascending=False)
```

## Most frequest value mode per column

```python
pd.Series({col: df_ads[col].mode().iloc[0] if not df_ads[col].mode().empty else np.nan 
           for col in df_ads.columns}, name="top_value")
```

## Group by page id

```python
# First, filter only numeric columns
numeric_cols = df_ads.select_dtypes(include="number").columns

# Then group and aggregate only those numeric columns
df_ads.groupby("page_id")[numeric_cols].agg(["count", "mean", "std", "min", "max"]).head()
```

## Group by page id and add id

```python
# 🔹 Select numeric columns first
numeric_cols = df_ads.select_dtypes(include="number").columns

# 🔹 Group by page_id + ad_id, then aggregate ONLY those numeric columns
grouped_page_ad = (
    df_ads.groupby(["page_id", "ad_id"])[numeric_cols]
          .agg(["count", "mean", "std", "min", "max"])
)

# Preview the first few groups
grouped_page_ad.head()
```

## Now we will load another dataset

```python
df_posts = pd.read_csv("2024_fb_posts_president_scored_anon.csv", low_memory=False)
df_posts.shape, df_posts.head(3)
```

## Basic descriptive stats

```python
df_posts.describe(include='all')
```

## Numeric Stats Only

```python
numeric_cols = df_posts.select_dtypes(include="number").columns
df_posts[numeric_cols].describe().T
```

## Unique counts in non-numeric columns

## Most frequent value (mode) per column

```python
pd.Series({
    col: df_posts[col].mode().iloc[0] if not df_posts[col].mode().empty else np.nan
    for col in df_posts.columns
}, name="top_value")
```

## Quick nunique snapshot

```python
# Unique-value counts for every column (top 20 shown)
df_posts.nunique().sort_values(ascending=False).head(20)
```

## Grouped Stats by Facebook Id and post Id separately 

```python
# Ensure column names are cleaned and numeric columns selected
df_posts.columns = df_posts.columns.str.strip().str.lower().str.replace(" ", "_")
numeric_cols = df_posts.select_dtypes(include="number").columns

grouped_fbid = (
    df_posts.groupby("facebook_id")[numeric_cols]
            .agg(["count", "mean", "std", "min", "max"])
)
grouped_fbid.head()
```

```python
# Select numeric columns
numeric_cols = df_posts.select_dtypes(include="number").columns

# Group by post_id and compute summary statistics
grouped_by_postid = (
    df_posts
    .groupby("post_id")[numeric_cols]
    .agg(["count", "mean", "std", "min", "max"])
)

# Display the first few rows
grouped_by_postid.head()
```

## Grouped Stats by Facebook Id and Post Id and page_category and post_id

```python
# Identify numeric columns only
numeric_cols = df_posts.select_dtypes(include="number").columns

# Group by page_category and facebook_id (page ID surrogate)
grouped_pagecat_fbid = (
    df_posts
    .groupby(["page_category", "facebook_id"])[numeric_cols]
    .agg(["count", "mean", "std", "min", "max"])
)

# Display first few rows
grouped_pagecat_fbid.head()
```

```python
# Select numeric columns
numeric_cols = df_posts.select_dtypes(include="number").columns

# Group by page_category and post_id
grouped_pagecat_postid = (
    df_posts
    .groupby(["page_category", "post_id"])[numeric_cols]
    .agg(["count", "mean", "std", "min", "max"])
)

# Display the first few rows
grouped_pagecat_postid.head()
```

## Now we will use another dataset

```python
import pandas as pd

# Load the Twitter posts data
df_tw = pd.read_csv("2024_tw_posts_president_scored_anon.csv")

# Quick look
df_tw.head()
```

## Basic Descriptive Stats

```python
df_tw.describe(include='all').T
```

## Numeric Stats

```python
numeric_cols_tw = df_tw.select_dtypes(include='number').columns.tolist()
numeric_cols_tw
```

## Unique Counts in non-numeric columns

```python
non_numeric_cols_tw = df_tw.select_dtypes(exclude='number').columns

df_tw[non_numeric_cols_tw].nunique().sort_values(ascending=False)
```

## Most Frequent value i.e. mode per column

```python
# Compute the mode (most frequent value) for each column in the Twitter dataset
top_values_tw = pd.Series({
    col: df_tw[col].mode().iloc[0] if not df_tw[col].mode().empty else np.nan
    for col in df_tw.columns
}, name="top_value")

# Display the top values
top_values_tw
```

## Quick nunique snapshot

```python
# Unique-value counts for every column in Twitter data (top 20 shown)
df_tw.nunique().sort_values(ascending=False).head(20)
```

## Descriptive stats by source

```python
numeric_cols_tw = df_tw.select_dtypes(include='number').columns.tolist()
grouped_source = (
    df_tw.groupby("source")[numeric_cols_tw]
         .agg(["count", "mean", "std", "min", "max"])
)
grouped_source.head()
```

## Descriptive stats by source and lang

```python
grouped_source_lang = (
    df_tw.groupby(["source", "lang"])[numeric_cols_tw]
         .agg(["count", "mean", "std", "min", "max"])
)
grouped_source_lang.head()
```
